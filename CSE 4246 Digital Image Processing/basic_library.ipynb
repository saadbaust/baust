```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 1. Read Color Image and Split into RGB Channels
image = cv2.imread('sample.png')  # Read color image

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.show()
```

```python
flip_horizontal = cv2.flip(image, 1)
flip_vertical = cv2.flip(image, 0)

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original")

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(flip_horizontal, cv2.COLOR_BGR2RGB))
plt.title("Hor flipped Image")

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(flip_vertical, cv2.COLOR_BGR2RGB))
plt.title("Ver flipped Image Matrix")
```

```python

bright_lib = cv2.convertScaleAbs(image, beta=100)  # Match manual increase


plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(bright_lib, cv2.COLOR_BGR2RGB))
plt.title("Library Brightness")
plt.show()
```

```python
invert_lib = cv2.bitwise_not(image)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(invert_lib, cv2.COLOR_BGR2RGB))
plt.title("Library Invert")
plt.show()
```
