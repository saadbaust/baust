```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 1. Read Normal Image and Convert to Grayscale using OpenCV
image = cv2.imread('sample.png')  # Read normal image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
print(gray_image.shape)

# Print grayscale matrix
print("Grayscale Image Matrix:")
print(gray_image)

plt.imshow(gray_image, cmap='gray')
plt.title("Converted Grayscale Image")
plt.show()
```

```python
def flip_horizontal(matrix):
    flipped = []
    for row in matrix:
        new_row = []
        for i in range(len(row) - 1, -1, -1):
            new_row.append(row[i])
        flipped.append(new_row)
    return flipped

def flip_vertical(matrix):
    flipped = []
    for i in range(len(matrix) - 1, -1, -1):
        flipped.append(matrix[i])
    return flipped

flip_horizontal = flip_horizontal(gray_image)
flip_vertical = flip_vertical(gray_image)

print("Hor flipped Image Matrix:")
print(np.array(flip_horizontal))
print("Ver flipped Image Matrix:")
print(np.array(flip_vertical))


plt.subplot(1, 3, 1)
plt.imshow(gray_image, cmap='gray')
plt.title("Original")

plt.subplot(1, 3, 2)
plt.imshow(flip_horizontal, cmap='gray')
plt.title("Hor flipped Image")

plt.subplot(1, 3, 3)
plt.imshow(flip_vertical, cmap='gray')
plt.title("Ver flipped Image Matrix")
```

```python
def brightness(matrix, value):
    result = []
    for row in matrix:
        new_row = []
        for pixel in row:
            pixel += value #increase
            # pixel -= value #decrease
            if pixel > 255:
                pixel = 255
            elif pixel < 0:
                pixel = 0
            new_row.append(pixel)
        result.append(new_row)
    return result

increased_brightness_manual = brightness(gray_image, 200)

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title("Original")
plt.subplot(1, 2, 2)
plt.imshow(increased_brightness_manual, cmap='gray')
plt.title("Manual Brightness")
plt.show()
```

```python
# 5. Invert Image Colors
def invert(matrix):
    result = []
    for row in matrix:
        new_row = []
        for pixel in row:
            new_row.append(255 - pixel)
        result.append(new_row)
    return result

inverted_manual = invert(gray_image)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title("Original")
plt.subplot(1, 2, 2)
plt.imshow(inverted_manual, cmap='gray')
plt.title("Manual Invert")
plt.show()
```

```python
b, g, r = cv2.split(image)  # Split into Blue, Green, Red channels

# Display the matrices
print("Red Channel Matrix:")
print(r)
print("Green Channel Matrix:")
print(g)
print("Blue Channel Matrix:")
print(b)

# Show the original image
plt.figure(figsize=(6, 4))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")
plt.show()

newimage=cv2.merge([r,g,b])
# Show the original image
plt.figure(figsize=(6, 4))
plt.imshow(newimage)
plt.title("Merged Image")
plt.axis("off")
plt.show()
```
